```python
from pyspark.sql.functions import year, month, col

dated_df = spark.sql("SELECT * FROM piotrt_demo.minifigs") \
	.withColumn("Year", year(col("LastModifiedDate"))) \
	.withColumn("Month", month(col("LastModifiedDate")))
    
display(dated_df)
```

```python
dated_df \
	.write \
	.partitionBy("Year", "Month") \
	.format("delta") \
	.option("path", "abfss://curated@tybuldatalakedemo.dfs.core.windows.net/Rebrickable/Minifigs_Synapse_Partitioned") \
    .saveAsTable("PiotrT_Demo.Minifigs_Synapse_Partitioned")
```
